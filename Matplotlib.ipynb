```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
```

## Scenario: Student Grades
Let's consider a scenario where we have the grades of three students in three subjects.

```python
# Student names
students = ['Alice', 'Bob', 'Charlie']

# Grades for three subjects (out of 100)
math_grades = [90, 85, 78]
english_grades = [80, 92, 88]
history_grades = [75, 87, 90]
phy_grades = [30, 50, 54]
sport_grades = [90, 90, 60]
```

***Basic Plotting***

```python
# 1. Basic Plotting
# Creating a simple line plot for math grades
plt.plot(students, math_grades, marker='o', label='Math Grades')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.title('Math Grades of Students')
plt.legend()
plt.show()
```

***Scatter Plots***

```python
# 2. Scatter Plots
# Creating a scatter plot for English and History grades
plt.scatter(students, english_grades, color='blue', label='English Grades', marker='s')
plt.scatter(students, history_grades, color='green', label='History Grades', marker='^')
plt.xlabel('Students')
plt.ylabel('Grades')
plt.title('English and History Grades of Students')
plt.legend()
plt.show()
```

***Bar Charts***

```python
# 3. Bar Charts
# Creating a bar chart for average grades
average_grades = np.mean([math_grades, english_grades, history_grades, phy_grades, sport_grades], axis=0)
plt.bar(students, average_grades, color='purple', label='Average Grades')
plt.xlabel('Students')
plt.ylabel('Average Grades')
plt.title('Average Grades of Students')
plt.legend()
plt.show()
```

***Pie Charts***

```python
# 4. Pie Charts
# Generating a pie chart for Math grades distribution
plt.pie(math_grades, labels=students, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Math Grades Distribution')
plt.show()
```

***Subplots***

```python
# 5. Subplots
# Creating subplots for each subject
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

axs[0].bar(students, math_grades, color='orange')
axs[0].set_title('Math Grades')

axs[1].bar(students, english_grades, color='blue')
axs[1].set_title('English Grades')

axs[2].bar(students, history_grades, color='green')
axs[2].set_title('History Grades')

plt.tight_layout(pad = 1)
plt.show()
```

***Loading the Iris dataset***

```python
# Loading the Iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
iris_df['target'] = iris_df['target'].astype(int)
# display(iris_df)
# print(iris_df['target'])
```

***Working with Real Data***

```python
# 8. Working with Real Data
# Plotting the sepal length and width of different iris species
plt.scatter(iris_df[iris_df['target'] == 0]['sepal length (cm)'], iris_df[iris_df['target'] == 0]['sepal width (cm)'], label='Setosa', marker='o', color='red')
plt.scatter(iris_df[iris_df['target'] == 1]['sepal length (cm)'], iris_df[iris_df['target'] == 1]['sepal width (cm)'], label='Versicolor', marker='s', color='green')
plt.scatter(iris_df[iris_df['target'] == 2]['sepal length (cm)'], iris_df[iris_df['target'] == 2]['sepal width (cm)'], label='Virginica', marker='^', color='blue')

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width of Iris Species')
plt.legend()
plt.show()
```
