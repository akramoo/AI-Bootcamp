# 10. Merging and Concatenating DataFrames


````python
# Combining DataFrames using merge
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [4, 5, 6], 'Name': ['David', 'Eva', 'Frank']})
merged_df = pd.merge(df1, df2, on='ID', how='inner')

# Concatenating DataFrames vertically and horizontally
concatenated_df = pd.concat([df1, df2], axis=0)



# Display the first dataFrame
display(df1)

# Display the second dataFrame
display(df2)

# Displaying the merged DataFrame
display(merged_df)

# Display the merged dataFrame
display(concatenated_df)
````

# 11. Time Series Data

````python
# Example time series data:
# link = https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv


# Working with date and time data
time_series_url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
time_series_df = pd.read_csv(time_series_url, parse_dates=['Month'], index_col='Month')

# Resampling and shifting time series data
resampled_data = time_series_df.resample('M').mean()
shifted_data = time_series_df.shift(periods=1)

# Diplaying resampled data
display(resampled_data)

# Diplaying shifted data
display(shifted_data)
````

# 15. Dealing with Datetime Data

````python
# Assuming you have a DataFrame named df with a column 'Date'
# Create a sample DataFrame for illustration purposes
data = {'Date': ['2023-01-01', '2023-02-15', '2023-03-20']}
df = pd.DataFrame(data)

# Extracting information from datetime columns
# Creating new datetime features
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Display the updated DataFrame
display(df)
````

````python
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming you have a DataFrame named time_seriesdf with a column 'Value'
# Create a sample DataFrame for illustration purposes
data = {'Date': pd.date_range(start='2023-01-01', periods=365),
        'Value': range(365)}
time_seriesdf = pd.DataFrame(data)
time_seriesdf.set_index('Date', inplace=True)

# Time Series Analysis
# Decomposing time series data
# Testing for stationarity

decomposition = seasonal_decompose(time_seriesdf['Value'])

# Display the decomposition results
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

print("Trend Component:")
display(trend)
print("\nSeasonal Component:")
display(seasonal)
print("\nResidual Component:")
display(residual)
````

````python
# Import necessary libraries
%pip install scikit-learn
import pandas as pd
from sklearn.datasets import load_iris
````

````python
# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['Target'] = iris.target
display(iris_df)
display(iris_df['Target'])
````

````python
# Display basic information about the dataset
print("Basic Information about the Iris Dataset:")
print(iris_df.info())
````

````python
# Check for missing values
print("\nChecking for Missing Values:")
print(iris_df.isnull().sum())
````

````python
# Explore basic statistics
print("\nBasic Statistics of the Iris Dataset:")
print(iris_df.describe())
````

````python
# Visualize the distribution of sepal lengths
# !pip install seaborn
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
plt.figure(figsize=(8, 6))
sns.histplot(data=iris_df, x='sepal length (cm)', kde=True, color='skyblue')
plt.title('Distribution of Sepal Lengths')
plt.show()
````

````python
# Identify the top 5 customers with the highest total spending
# (In this case, let's assume sepal length as spending for simplicity)
top_spending_customers = iris_df.nlargest(5, 'sepal length (cm)')
display(top_spending_customers)
````

````python
# Determine the most frequently purchased product
most_frequent_species = iris_df['Target'].value_counts().idxmax()
display(most_frequent_species)
````

````python
# Convert the 'sepal length (cm)' column to a datetime format (for illustration purposes)
iris_df['Timestamp'] = pd.to_datetime(iris_df['sepal length (cm)'], unit='s')
display(iris_df['Timestamp'])
````

````python
# Explore monthly trends in 'sepal length (cm)' (for illustration purposes)
monthly_trends = iris_df.resample('M', on='Timestamp')['sepal length (cm)'].mean()
display(monthly_trends)
````

````python
# Find the average 'sepal length (cm)' of each species
average_length_per_species = iris_df.groupby('Target')['sepal length (cm)'].mean()
display(average_length_per_species)
````

````python
# Identify the species with the highest total sepal length
highest_total_length_species = iris_df.groupby('Target')['sepal length (cm)'].sum().idxmax()
display(highest_total_length_species)
````

````python
# Create a new feature 'TotalCost' by multiplying 'sepal length (cm)' and 'Target'
iris_df['TotalCost'] = iris_df['sepal length (cm)'] * iris_df['Target']
display(iris_df['TotalCost'])
````

````python
# Analyze the correlation between 'sepal length (cm)' and 'TotalCost'
correlation = iris_df['sepal length (cm)'].corr(iris_df['TotalCost'])
display(correlation)
````
